## 作業3
#### 請寫一個函式用來計算 Mean Square Error
$ MSE = \frac{1}{n}\sum_{i=1}^{n}{(Y_i - \hat{Y}_i)^2} $

### Hint: [如何取平方](https://googoodesign.gitbooks.io/-ezpython/unit-1.html)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_squared_error(y, yb):
    mse = MSE =sum((y-yb)**2)/len(y)
    return mse
def mean_absolute_error(y, yb):
    mae = MAE =sum(abs(y-yb))/len(y)
    return mae
```

```python
w = 3
b = 0.5

x_lin = np.linspace(0, 100, 101)  #0到100間分成101個數

y = (x_lin + np.random.randn(101) * 5) * w + b  #randn函數生成101個數值 斜率和截距為w/b 當作是實際的值

plt.plot(x_lin, y, 'b.', label = 'data points') #b.劃出藍色的點點
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()
```

```python
y_hat = x_lin * w + b  #y_hat是在線上的點 理論上的值 就是預測值
plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
# 執行 Function, 確認有沒有正常執行
MSE = mean_squared_error(y, y_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
print("The Mean absolute error is %.3f" % (MAE))
```
